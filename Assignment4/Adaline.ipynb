```python
import pandas as pd
import numpy as np

import seaborn as sns
sns.set()
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
```

```python
df = pd.read_csv('dataset.csv',sep=';')
df = df.rename(columns=df.iloc[0]).drop(df.index[0])
df = df.drop(columns = 'ID')
df
```

```python
df['Motorway'] = df.Motorway.apply(lambda x:1 if x =='A1' else 0)
df
```

```python
df.columns = [c.replace(' ','_').lower() for c in df.columns]
df
```

```python
df['tree_frog'] = df.tree_frog.apply(lambda x: -1 if x == '0' else x)
df['tree_frog'] = pd.to_numeric(df['tree_frog'])
df
```

```python
df.columns
```

# X and y with minmax scaling

```python
features = ['motorway', 'sr', 'nr', 'tr', 'vr', 'sur1', 'sur2', 'sur3', 'ur', 'fr', 'or', 'rr', 'br', 'mr', 'cr']
X = df[features]
y = df['tree_frog'].to_numpy()
X = minmax.fit_transform(X)
X = np.c_[np.ones(X.shape[0]),X]
```

# ML functions

```python
def J(X,y,w):
    yhat = X@w
    return np.sum((y-yhat)**2/2)

def Jgrad(X,y,w):
    """The gradient of J with respect to w"""
    yhat = (X@w)
    difference = (y-yhat).reshape(X.shape[0],1)
    return np.sum(difference*(X),axis=0)/X.shape[0]

def fit_adeline(X,y,max_epochs=50,eta=0.1):
    """Takes X and y and returns a trained w"""
    n,d = X.shape
    w = np.random.randn(d)
    errors = []
    W_history = []
    for epoch in range(max_epochs):
        W_history.append(w.copy())
        if epoch%(epoch//10+1) == 0:
            print(w,"error=",J(X,y,w))
        errors.append(J(X,y,w))
        dw = Jgrad(X,y,w)
        w += eta*dw
    return w,errors,W_history

def fit_adeline_sgd(X,y,max_epochs=50,eta=0.1):
    """Takes X and y and returns a trained w"""
    n,d = X.shape
    w = np.random.randn(d)
    errors = []
    W_history = []
    for epoch in range(max_epochs):
        W_history.append(w.copy())
        if epoch%(epoch//10+1) == 0:
            print(w,"error=",J(X,y,w))
        errors.append(J(X,y,w))
        for x,y_i in zip(X,y):
            yhat = w.T@x
            dw = eta*(y_i-yhat)*x
            w += dw
    return w,errors,W_history

def fit_adeline_sgd_batch(X,y,max_epochs=50,eta=0.1,batch_size = 32):
    """Takes X and y and returns a trained w"""
    n,d = X.shape
    w = np.random.randn(d)
    errors = []
    W_history = []
    Xpartition = np.array_split(X,X.shape[0]//batch_size)
    ypartition = np.array_split(y,len(y)//batch_size)
    for epoch in range(max_epochs):
        W_history.append(w.copy())
        if epoch%(epoch//10+1) == 0:
            print(w,"error=",J(X,y,w))
        errors.append(J(X,y,w))
        for x_i,y_i in zip(Xpartition,ypartition):
            dw = Jgrad(x_i,y_i,w)
            w += eta*dw
    return w,errors,W_history
```

# Vanilla adaline

```python
w,errors,W_history= fit_adeline(X,y,max_epochs=5000,eta=0.1)
W_history = np.array(W_history)
w
```

```python
plt.plot(errors[:15])
plt.title(r"Error metric $J(w)$")
plt.xlabel("Epochs")
plt.ylabel(r"$J(w)$")
plt.show()
```

# stochastic gradient adaline

```python
w,errors,W_history= fit_adeline_sgd(X,y,max_epochs=5000,eta=0.1)
W_history = np.array(W_history)
w
```

```python
plt.plot(errors[:15])
plt.title(r"Error metric sgd")
plt.xlabel("Epochs")
plt.ylabel(r"$J(w)$")
plt.show()
```

# adding functions to shuffle rows

```python
def shuffle(X, y):
    p = np.random.permutation(len(y))
    return X[p], y[p]
def fit_adeline_sgd(X,y,max_epochs=50,eta=0.1):
    """Takes X and y and returns a trained w"""
    n,d = X.shape
    w = np.random.randn(d)
    errors = []
    W_history = []
    for epoch in range(max_epochs):
        W_history.append(w.copy())
        if epoch%(epoch//10+1) == 0:
            print(w,"error=",J(X,y,w))
        errors.append(J(X,y,w))
        for x,y_i in zip(X,y):
            yhat = w.T@x
            dw = eta*(y_i-yhat)*x
            w += dw
        X,y = shuffle(X,y)
    return w,errors,W_history
```
