### Universal functions

Numpy makes heavy use of "vectorized" operations. 

Vectorized operations happen on many items in an arraylike object simultaneously.

We have already seen this with boolean indexing, scalar addition, and "broadcast" (ie itemwise) multiplication of numpy arrays.

The same is true for applying functions to arrays.   

For example suppose you want to apply the `sine` function to every element in some array.



```python
import numpy as np

x = np.linspace(-8*np.pi,8*np.pi,200)
x
```

```python
##  The numpy universal function "sine".

y = np.sin(x)
y
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.show()
```

```python
z = np.abs(y)

plt.plot(x,z)

plt.show()
```
